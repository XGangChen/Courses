<p>Using the Sobel edge detector, calculate the <b>gradient direction</b> and <b>gradient magnitude</b>
of the center pixel (pixel[ 4, 4 ]=212 ).</p>

<div>
Gradient direction: ENTER VALE<br>
Gradient magnitude: ENTER VALUE

```python
# Define the Sobel kernels
Gx_kernel = np.array([[-1, 0, 1], 
                      [-2, 0, 2], 
                      [-1, 0, 1]])

Gy_kernel = np.array([[-1, -2, -1], 
                      [ 0,  0,  0], 
                      [ 1,  2,  1]])

# Define the 3x3 neighborhood around pixel (4, 4)
neighborhood = np.array([[253, 212, 19], 
                         [155, 214, 160], 
                         [91,  212, 244]])

# Convolve with Gx and Gy
Gx = np.sum(Gx_kernel * neighborhood)
Gy = np.sum(Gy_kernel * neighborhood)

# Calculate gradient magnitude and direction
gradient_magnitude = np.sqrt(Gx**2 + Gy**2)
gradient_direction = np.degrees(np.arctan2(Gy, Gx))  # Convert to degrees

gradient_magnitude, gradient_direction
```

```python
# Define the 3x3 neighborhood again for clarity
neighborhood = np.array([[253, 212, 19], 
                         [155, 214, 160], 
                         [91,  212, 244]])

# Function to calculate Gx, Gy, magnitude, and angle for a pixel
def calculate_gradients(neighborhood, Gx_kernel, Gy_kernel):
    Gx = np.zeros_like(neighborhood, dtype=float)
    Gy = np.zeros_like(neighborhood, dtype=float)
    magnitude = np.zeros_like(neighborhood, dtype=float)
    angle = np.zeros_like(neighborhood, dtype=float)

    for i in range(1, neighborhood.shape[0] - 1):
        for j in range(1, neighborhood.shape[1] - 1):
            # Get local 3x3 patch
            local_patch = neighborhood[i-1:i+2, j-1:j+2]
            Gx_val = np.sum(local_patch * Gx_kernel)
            Gy_val = np.sum(local_patch * Gy_kernel)
            Gx[i, j] = Gx_val
            Gy[i, j] = Gy_val
            magnitude[i, j] = np.sqrt(Gx_val**2 + Gy_val**2)
            angle[i, j] = np.abs(np.degrees(np.arctan2(Gy_val, Gx_val))) % 180

    return magnitude, angle

# Compute gradients
magnitude, angle = calculate_gradients(neighborhood, Gx_kernel, Gy_kernel)

# Define bins for HOG (9 bins: 20-degree steps between 0 to 180 degrees)
bins = np.arange(0, 181, 20)
histogram = np.zeros(len(bins) - 1)

# Accumulate magnitudes into bins based on angle
for i in range(1, neighborhood.shape[0] - 1):
    for j in range(1, neighborhood.shape[1] - 1):
        angle_value = angle[i, j]
        mag_value = magnitude[i, j]
        # Find the appropriate bin and add the magnitude
        for k in range(len(bins) - 1):
            if bins[k] <= angle_value < bins[k + 1]:
                histogram[k] += mag_value
                break

histogram
```

1.Compute the gradient magnitude and direction for each pixel using Sobel filters Gx and Gy.
2.Map the gradient directions to bins of 20-degree increments (0°–20°, 20°–40°, ..., 160°–180°).
3.Accumulate weighted values (gradient magnitudes) into each bin based on their respective gradient directions.

```python
# Generator function
def generator(size, _theta=10.0 / 180.0 * math.pi):
    img = np.zeros(size, dtype=np.uint8)
    n = random.randint(5, 15)
    for k in range(n):
        l = random.randint(size[0] // 2, size[0])
        d = 0
        x, y = random.randint(0, size[0] - 1), random.randint(0, size[1] - 1)
        theta = _theta
        while d < l:
            if random.random() < 0.2:
                theta = random.gauss(_theta, math.pi / 5)
            x, y = x + math.cos(theta), y - math.sin(theta)
            if (0 <= x < len(img[0])) and (0 <= y < len(img)):
                img[int(y), int(x)] = 255
            d += 1
    return img

# Parameters for HOG
def extract_hog_features(image):
    win_size = (128, 128)
    block_size = (64, 64)
    block_stride = (32, 32)
    cell_size = (32, 32)
    nbins = 9
    
    hog = cv2.HOGDescriptor(win_size, block_size, block_stride, cell_size, nbins)
    image_resized = cv2.resize(image, win_size)
    features = hog.compute(image_resized)
    return features.flatten()

# Visualization function
def visualize_grass_types(type1_images, type2_images, size=(64, 64)):
    fig, axes = plt.subplots(2, 5, figsize=(15, 5))
    fig.suptitle("Grass Type 1 - generator((64, 64),0.17)", fontsize=12)
    for i, ax in enumerate(axes.flatten()[:10]):
        if i < len(type1_images):
            ax.imshow(type1_images[i], cmap='inferno')
        ax.axis('off')

    fig, axes = plt.subplots(2, 5, figsize=(15, 5))
    fig.suptitle("Grass Type 2 - generator((64, 64),1.22)", fontsize=12)
    for i, ax in enumerate(axes.flatten()[:10]):
        if i < len(type2_images):
            ax.imshow(type2_images[i], cmap='inferno')
        ax.axis('off')

# Generate training and testing data
train_images_type1 = [generator((128, 128), _theta=10.0 / 180.0 * math.pi) for _ in range(40)]
train_images_type2 = [generator((128, 128), _theta=30.0 / 180.0 * math.pi) for _ in range(40)]
test_images_type1 = [generator((128, 128), _theta=10.0 / 180.0 * math.pi) for _ in range(10)]
test_images_type2 = [generator((128, 128), _theta=30.0 / 180.0 * math.pi) for _ in range(10)]

# Extract HOG features for training images
hog_features_type1 = np.array([extract_hog_features(img) for img in train_images_type1])
hog_features_type2 = np.array([extract_hog_features(img) for img in train_images_type2])

# Compute centroids
centroid_type1 = np.mean(hog_features_type1, axis=0)
centroid_type2 = np.mean(hog_features_type2, axis=0)

# Classification function
def classify_image(image):
    features = extract_hog_features(image)
    dist1 = np.linalg.norm(features - centroid_type1)
    dist2 = np.linalg.norm(features - centroid_type2)
    return 1 if dist1 < dist2 else 2

# Testing
true_labels = [1] * 10 + [2] * 10
predicted_labels = [classify_image(img) for img in test_images_type1] + [classify_image(img) for img in test_images_type2]

# Accuracy
try:
    accuracy = accuracy_score(true_labels, predicted_labels)
except NameError:
    from sklearn.metrics import accuracy_score
    accuracy = accuracy_score(true_labels, predicted_labels)

# Visualization of test images
test_images_type1_visual = [generator((64, 64), _theta=0.17) for _ in range(10)]
test_images_type2_visual = [generator((64, 64), _theta=1.22) for _ in range(10)]
visualize_grass_types(test_images_type1_visual, test_images_type2_visual)
plt.show()

print("Classification Accuracy:", accuracy)
```
